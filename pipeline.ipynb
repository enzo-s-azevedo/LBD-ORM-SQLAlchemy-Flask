# Pipeline Interativo

Execute as células abaixo na ordem desejada. Cada bloco de texto descreve sucintamente o código abaixo.

```python
# Setup: importe a app e os modelos (execute uma vez)
from app import create_app
from models import db, Usuario, Musica, Playlist, MusicaPlaylist
from relacionamento import criar_playlist, adicionar_musica_playlist, remover_musica_playlist

app = create_app()
print('App carregada')
```

# Criando Playlist
Cria a playlist `Sertanejos do Josue` para o usuário `Josue`. O código é idempotente: não criará duplicatas.

```python
with app.app_context():
    josue = Usuario.query.filter_by(username='Josue').first()
    if not josue:
        print('Usuário 
 não encontrado. Rode `python3 popular_bd.py` primeiro.')
    else:
        existente = Playlist.query.filter_by(nome='Sertanejos do Josue', usuario_id=josue.id).first()
        if existente:
            print('Playlist já existe:', existente)
        else:
            p = criar_playlist(josue.id, 'Sertanejos do Josue')
            print('Playlist criada:', p)
```

# Adicionando música
Adiciona `Stairway to Heaven` na playlist `Sertanejos do Josue` (se existir).

```python
with app.app_context():
    josue = Usuario.query.filter_by(username='Josue').first()
    if not josue:
        print('Usuário 
 não encontrado. Rode `python3 popular_bd.py` primeiro.')
    else:
        p = Playlist.query.filter_by(nome='Sertanejos do Josue', usuario_id=josue.id).first()
        if not p:
            print('Playlist não encontrada; execute a célula Criando Playlist primeiro.')
        else:
            stairway = Musica.query.filter_by(titulo='Stairway to Heaven').first()
            if not stairway:
                print("Música 'Stairway to Heaven' não encontrada; rode `python3 popular_bd.py` primeiro.")
            else:
                mp = adicionar_musica_playlist(josue.id, p.playlist_id, stairway.id, None)
                print('Música adicionada:', mp)
```

# Removendo música
Remove `Stairway to Heaven` da playlist `Sertanejos do Josue` (se presente).

```python
with app.app_context():
    josue = Usuario.query.filter_by(username='Josue').first()
    if not josue:
        print('Usuário 
 não encontrado.')
    else:
        p = Playlist.query.filter_by(nome='Sertanejos do Josue', usuario_id=josue.id).first()
        if not p:
            print('Playlist não encontrada; nada a remover.')
        else:
            stairway = Musica.query.filter_by(titulo='Stairway to Heaven').first()
            if not stairway:
                print("Música 'Stairway to Heaven' não encontrada; nada a remover.")
            else:
                removed = remover_musica_playlist(josue.id, p.playlist_id, stairway.id)
                print('Removido:', removed)
```
